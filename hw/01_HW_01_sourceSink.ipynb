# Источник и сток

Потенциальная аэродинаимка.
Упрощения:
- стационарные течеиния;
- несжимаемые течения (скорость много меньше скорости звука);
- невязкая жидкость;
- без завихренности

##Немного формул:
---
По теореме Стокса циркуляцию
$$\Gamma = \oint \mathbf{v}\cdot d\vec{l}$$
можно записать как поверхностный интеграл завихренности (которая есть $\omega = \mathrm{rot}\, \mathbf{v}$ или $\omega = \nabla \times \mathbf{v}$):

$$\oint \mathbf{v}\cdot d\vec{l} = \int \int_s \omega \cdot \vec{n}~ds$$

Для безвихревых течений $\omega = 0$, и интеграл скорости по любой кривой AB равен интегралу по любой кривой BA. Раскрываем скалярное произведение:

$$\int_A^B \mathbf{v} \cdot d\vec{l} = \int_A^B u~dx+v~dy+w~dz$$

В безвихревом течении значение полученного интграла не зависит от пути интегрирования, что зозначает, что $u~dx+v~dy+w~dz$ есть полный дифференциал $d~ \phi$:

$$u=\frac{\partial \phi}{\partial x}, \quad v=\frac{\partial \phi}{\partial y}, \quad w=\frac{\partial \phi}{\partial z}$$

или $\mathbf{v} = \nabla\phi$. Вместе с уравнением неразрывности для несжимаемой жидкости $\nabla \cdot \mathbf{v} = 0$ получим _уравнение Лапласа_:
$$\nabla^2\phi = 0$$

То есть любое решение уравнения лапласа является потенциальным течением.

## Сетка для линий тока

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

Создаем равномерную сетку, на которой будем рисовать

```python
N = 51
x_start,x_end = -2.,2
y_start,y_end = -1.,1.
x = np.linspace(x_start,x_end,N)
y = np.linspace(y_start,y_end,N)

print x

X,Y = np.meshgrid(x,y) 
```

```python
%matplotlib inline

size= 10
plt.figure(figsize = (size,(y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x',fontsize = 16)
plt.ylabel('y',fontsize = 16)
plt.xlim(x_start,x_end)
plt.ylim(y_start,y_end)
plt.scatter(X,Y,s=10,color = '#CD2305',marker = 'o', linewidth = 0)
```

## Функция тока и потенциал для источника

$\sigma$ $-$ интенсивность (секундный расход)

Комплексный потенциал $-$ $\chi(z) = \frac{\sigma}{2\pi}\ln~z$

Потенциал  $-$ $\phi = \frac{\sigma}{2\pi}\ln~r$

Функция тока $-$ $\psi = \frac{\sigma}{2\pi}~\theta$

Компоненты скорости в декартовых координатах:
$$u = \frac{\partial\psi}{\partial y} = \frac{\sigma}{2\pi}\frac{x-x_\text{source}}{\left(x-x_\text{source}\right)^2+\left(y-y_\text{source}\right)^2}$$


$$v = -\frac{\partial\psi}{\partial x} = \frac{\sigma}{2\pi}\frac{y-y_\text{source}}{\left(x-x_\text{source}\right)^2+\left(y-y_\text{source}\right)^2}$$

$\left(x_\text{source},y_\text{source}\right)$  $-$ координаты источника

$\theta = \arctan(\frac{x}{y})$

## Линии тока для источника

```python
sigma_source = 5.0
x_source,y_source = -1.0,0.

#вычисляем массив скоростей
u_source = sigma_source/(2*math.pi) * (X-x_source)/((X-x_source)**2 + (Y-y_source)**2)
v_source = sigma_source/(2*math.pi) * (Y-y_source)/((X-x_source)**2 + (Y-y_source)**2)
```

```python
size = 10
plt.figure(figsize = (size,(y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x',fontsize = 16)
plt.ylabel('y',fontsize = 16)
plt.xlim(x_start,x_end)
plt.ylim(y_start,y_end)
plt.streamplot(X,Y,u_source,v_source,density=2,linewidth=1, arrowsize=2,arrowstyle='->')
plt.scatter(x_source,y_source,s=80,color = '#CD2305',marker = 'o', linewidth = 0);
```

## Линии тока для стока

```python
sigma_sink = -5.0
x_sink,y_sink = 1.0,0.

#вычисляем массив скоростей
u_sink = sigma_sink/(2*math.pi) * (X-x_sink)/((X-x_sink)**2 + (Y-y_sink)**2)
v_sink = sigma_sink/(2*math.pi) * (Y-y_sink)/((X-x_sink)**2 + (Y-y_sink)**2)
```

```python
size = 10
plt.figure(figsize = (size,(y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x',fontsize = 16)
plt.ylabel('y',fontsize = 16)
plt.xlim(x_start,x_end)
plt.ylim(y_start,y_end)
plt.streamplot(X,Y,u_sink,v_sink,density=2,linewidth=1, arrowsize=2,arrowstyle='->')
plt.scatter(x_sink,y_sink,s=80,color = '#CD2305',marker = 'o', linewidth = 0);
```

## Линии тока для пары

```python
u_pair = u_source + u_sink
v_pair = v_source + v_sink

size = 10
plt.figure(figsize = (size,(y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x',fontsize = 16)
plt.ylabel('y',fontsize = 16)
plt.xlim(x_start,x_end)
plt.ylim(y_start,y_end)
plt.streamplot(X,Y,u_pair,v_pair,density=2,linewidth=1, arrowsize=2,arrowstyle='->')
plt.scatter([x_source,x_sink],[y_source,y_sink],s=80,color = '#CD2305',marker = 'o', linewidth = 0);
```

## Потенциал для источника

```python
phi_source = sigma_source/(2*math.pi)*np.log((
(X-x_source)**2 + (Y-y_source)**2)**0.5)
phi_sink = sigma_sink/(2*math.pi)*np.log((
(X-x_sink)**2 + (Y-y_sink)**2)**0.5)
phi_pair = phi_source + phi_sink
size = 10
plt.figure(figsize = (size,(y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x',fontsize = 16)
plt.ylabel('y',fontsize = 16)
plt.xlim(x_start,x_end)
plt.ylim(y_start,y_end)
plt.streamplot(X,Y,u_pair,v_pair,density=2,linewidth=1, arrowsize=2,arrowstyle='->')
plt.contour(X,Y,phi_pair,colors='k',linewidth = 1,linstyles = 'solid')
plt.scatter([x_source,x_sink],[y_source,y_sink],s=80,color = '#CD2305',marker = 'o', linewidth = 0);
```
